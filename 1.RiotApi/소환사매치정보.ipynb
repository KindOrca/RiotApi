```python
from confluent_kafka import Producer
import json
import pandas as pd

# Kafka 프로듀서 설정
conf = {'bootstrap.servers': "localhost:9092"}
producer = Producer(**conf)

# CSV 파일을 읽어와서 Kafka로 전송
csv_filename = 'matches.csv'
csv_data = pd.read_csv(csv_filename)

def acked(err, msg):
    if err is not None:
        print(f"Failed to deliver message: {err}")
    else:
        print(f"Message produced: {msg.value()}")

for index, row in csv_data.iterrows():
    message = json.dumps(row.to_dict()).encode('utf-8')
    producer.produce('riot-matches', value=message, callback=acked)

# 전송된 모든 메시지가 브로커에 전달되도록 보장
producer.flush()
```

```python
from confluent_kafka import Consumer, KafkaError
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, BooleanType
import json

KAFKA_SERVER = 'localhost:9092'
DETAILS_TOPIC = 'match_details'
HDFS_PATH = 'hdfs:///user/match_data/'

def consume_from_kafka(topic, group_id):
    consumer_conf = {
        'bootstrap.servers': KAFKA_SERVER,
        'group.id': group_id,
        'auto.offset.reset': 'earliest'
    }
    consumer = Consumer(consumer_conf)
    consumer.subscribe([topic])
    return consumer

def extract_relevant_data(match_data):
    relevant_data = []
    seen_match_ids = set()
    
    for match in match_data:
        match_id = match.get('metadata', {}).get('matchId')
        if match_id in seen_match_ids:
            continue
        seen_match_ids.add(match_id)
        
        info = match.get('info', {})
        participants = info.get('participants', [])
        
        for participant in participants:
            player_data = {
                'matchId': match_id,
                'gameDuration': info.get('gameDuration'),
                # 'gameMode': info.get('gameMode'),
                'championId': participant.get('championId'),
                'championName': participant.get('championName'),
                'kills': participant.get('kills'),
                'deaths': participant.get('deaths'),
                'assists': participant.get('assists'),
                'win': participant.get('win')
            }
            relevant_data.append(player_data)
    return relevant_data

def save_to_hdfs_parquet(data, path):
    spark = SparkSession.builder.appName('MatchDataSaver').getOrCreate()
    
    # Define the schema
    schema = StructType([
        StructField("matchId", StringType(), True),
        StructField("gameDuration", IntegerType(), True),
        StructField("gameMode", StringType(), True),
        StructField("championId", IntegerType(), True),
        StructField("championName", StringType(), True),
        StructField("kills", IntegerType(), True),
        StructField("deaths", IntegerType(), True),
        StructField("assists", IntegerType(), True),
        StructField("win", BooleanType(), True)
    ])
    
    df = spark.createDataFrame(data, schema=schema)
    df.write.mode('overwrite').parquet(path)

details_consumer = consume_from_kafka(DETAILS_TOPIC, 'match_details_group')
details_data = []
print(1)
try:
    while True:
        msg = details_consumer.poll(1.0)
        print(1)
        if msg is None:
            continue
        if msg.error():
            if msg.error().code() == KafkaError._PARTITION_EOF:
                continue
            else:
                print(msg.error())
                break
        print(1)
        details_data.append(msg.value().decode('utf-8'))
        if len(details_data) >= 3:
            match_data_json = [json.loads(data) for data in details_data]
            relevant_data = extract_relevant_data(match_data_json)
            save_to_hdfs_parquet(relevant_data, f'{HDFS_PATH}/match_details_parquet')
            details_data = []
except KeyboardInterrupt:
    pass
finally:
    details_consumer.close()

# 남은 데이터 저장
if details_data:
    match_data_json = [json.loads(data) for data in details_data]
    relevant_data = extract_relevant_data(match_data_json)
    save_to_hdfs_parquet(relevant_data, f'{HDFS_PATH}/match_details_parquet')
```

```python
from confluent_kafka import Consumer, KafkaError
from hdfs import InsecureClient
import json

# Kafka 컨슈머 설정
conf = {
    'bootstrap.servers': "localhost:9092",
    'group.id': "hadoop-group",
    'auto.offset.reset': 'earliest'
}
consumer = Consumer(**conf)
consumer.subscribe(['riot-matches'])

# HDFS 클라이언트 설정
hdfs_client = InsecureClient('http://localhost:9870', user='hadoop')
hdfs_file_path = '/user/User/test_file/test.json'

# HDFS에 데이터를 저장할 함수
def write_to_hdfs(data):
    with hdfs_client.write(hdfs_file_path, encoding='utf-8', overwrite=True) as writer:
        writer.write(data)

# Kafka에서 메시지 수신 및 HDFS에 저장
try:
    while True:
        msg = consumer.poll(1.0)  # 1초 대기
        if msg is None:
            continue
        if msg.error():
            if msg.error().code() == KafkaError._PARTITION_EOF:
                # 모든 메시지를 읽은 상태
                continue
            else:
                print(msg.error())
                break

        message = msg.value().decode('utf-8')
        print(f"Received message: {message}")
        # HDFS에 저장
        write_to_hdfs(message)
finally:
    consumer.close()
```
